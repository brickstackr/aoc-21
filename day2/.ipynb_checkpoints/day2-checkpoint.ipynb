```python
#imports
import re
```

```python
with open("data.txt") as f:

  data = f.read()
```

```python
print(data)
```

```python
chunks = re.split('\n', data)
print(chunks)
```

```python
length = len(chunks)
print(length)
```

```python
counter = 0
depthcounter = 0
for i in range(length):
    a = chunks[i].split(' ')
    b = a[0]
    c = int(a[1])
    d = (b,c)
    if b == "forward":
        counter = counter + c
    if b == "up":
        depthcounter = depthcounter - c
    if b == "down":
        depthcounter = depthcounter + c
print(counter)
print(depthcounter)
```

```python
total = counter * depthcounter
print(total)
```

```python
# part2
aim = 0
counter = 0
depthcounter = 0
for i in range(length):
    a = chunks[i].split(' ')
    b = a[0]
    c = int(a[1])
    d = (b,c)
    if b == "down":
        aim = aim + c
    if b == "up":
        aim = aim - c
    if b == "forward":
        counter = counter + c
        depthcounter = depthcounter + aim*c
print(counter)
print(depthcounter)
```

```python
total2 = counter * depthcounter
print(total2)
```
