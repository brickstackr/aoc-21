```python
#imports
import re
```

```python
with open("data.txt") as f:

  data = f.read()
```

```python
print(data)
```

```python
chunks = re.split('\n', data)
print(chunks)
```

```python
length = len(chunks)
print(length)
```

```python
counter = 0
for i in range(length):
    a = int(chunks[i-1])
    b = int(chunks[i])
    if b > a:
        counter = counter + 1
print(counter)
```

```python
counter = 0
for i in range(length):
    a = int(chunks[i-3])
    b = int(chunks[i-2])
    c = int(chunks[i-1])
    d = int(chunks[i])
    sum1 = a+b+c
    sum2 = b+c+d
    if sum2 > sum1:
        counter = counter + 1
print(counter)
```
